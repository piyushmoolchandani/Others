```python
import requests
import json
```

```python
def get_data(url):
    url = "https://o136z8hk40.execute-api.us-east-1.amazonaws.com/dev/get-list-of-conferences"
    text = requests.get(url).text
    jsonData = json.loads(text) 
    return jsonData

def get_paid(jsonData):
    paidData = jsonData["paid"]
    keys = paidData[0].keys()
    ans = []
    for i in paidData:
        ans.append(", ".join([str(i[j]) for j in keys]))
    return ans;

def get_free(jsonData):
    freeData = jsonData["free"]
    keys = freeData[0].keys()
    ans = []
    for i in freeData:
        ans.append(", ".join([str(i[j]) for j in keys]))
    return ans;
```

```python
url = "https://o136z8hk40.execute-api.us-east-1.amazonaws.com/dev/get-list-of-conferences"
data = get_data(url)
paidData = get_paid(data)
freeData = get_free(data)
```

### Part 1: Printing

```python
print("Printing data in human readable format: ")
print("Paid: ")
for i in range(len(paidData)):
    print("Entry " + str(i + 1) + ":", paidData[i])
    print("\n----------------------------------------------------------------------------------------------------------\n")
print("Free: ")
for i in range(len(freeData)):
    print("Entry " + str(i + 1) + ":", freeData[i])
    print("\n----------------------------------------------------------------------------------------------------------\n")
```

### Part 2: Exactly Same

```python
# sort the entries
sortedFreeData = sorted(freeData)
sortedPaidData = sorted(paidData)

# To check for exactly same entries after sorting we only need to check with previous entry so printing the groups of non-similar entries
print("Paid: \n")
print(sortedPaidData[0], "\n")
for i in range(1, len(sortedPaidData)):
    if (sortedPaidData[i] == sortedPaidData[i - 1]):
        print("\n", sortedPaidData[i])
    else:
        print("\n-----------------------------------------------------------------------------------------------------------------------------\n\n", sortedPaidData[i])

print("Free: \n")
print(sortedFreeData[0], "\n")
for i in range(1, len(sortedFreeData)):
    if (sortedFreeData[i] == sortedFreeData[i - 1]):
        print("\n", sortedFreeData[i])
    else:
        print("\n-----------------------------------------------------------------------------------------------------------------------------\n\n", sortedFreeData[i])
```
